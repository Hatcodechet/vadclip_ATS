```python
import cv2

video_path = "/Users/hunghehe2205/Projects/VAD/Capstone/video/RoadAccidents074_x264.mp4"

cap = cv2.VideoCapture(video_path)

# Lấy FPS
fps = cap.get(cv2.CAP_PROP_FPS)

# Lấy tổng số frame
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

print("FPS:", fps)
print("Total frames:", frame_count)

cap.release()
```
